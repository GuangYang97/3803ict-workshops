```python
import numpy as np
import pandas as pd
import matplotlib.pyplot as plt
import seaborn as sns

# Read data from the CSV file
data = pd.read_csv("job-market(2).csv")
```

```python
# Count the number of jobs per location
job_counts = data["Location"].value_counts()

# Create the bar plot
plt.figure(figsize=(12, 6))
sns.barplot(x=job_counts.index, y=job_counts.values)
plt.xlabel("Location")
plt.ylabel("Number of Jobs")
plt.title("Number of Jobs by Location")
plt.xticks(rotation=90)  # Rotate x-axis labels for better readability

# Show the plot
plt.show()
```

```python
# Calculate the average salary for each job post
data["AverageSalary"] = (data["LowestSalary"] + data["HighestSalary"]) / 2

# Create the histogram
plt.figure(figsize=(12, 6))
plt.hist(data["AverageSalary"], bins=20, edgecolor="black")
plt.xlabel("Average Salary")
plt.ylabel("Number of Job Posts")
plt.title("Job Posts by Salary Range")

# Show the plot
plt.show()

```

```python
# Count the number of jobs per city
job_counts = data["Location"].value_counts()

# Plot the pie chart
plt.figure(figsize=(12, 6))
plt.pie(job_counts.values, labels=job_counts.index, autopct="%.1f%%", startangle=90)
plt.axis("equal")  # Equal aspect ratio ensures that pie is drawn as a circle
plt.title("Job Market Share by City")

# Show the plot
plt.show()
```
